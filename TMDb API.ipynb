Note: Do not use the search/movies API for finding genres.

```python
url = 'https://api.themoviedb.org/3/movie/top_rated'
params = {'language':'en-US', 'region':'US'}
r = requests.get(url, headers=headers, params=params)
py_data = r.json()
py_data
```

```python
title = []
genre = []
for i in py_data['results']:
    if (i['original_language']=='en' and len(title)!=10):
        title.append(i['title'])
        genre.append(i['genre_ids'])
```

```python
genre_url = 'https://api.themoviedb.org/3/genre/movie/list'
r1 = requests.get(genre_url, headers=headers)
py = r1.json()
```

```python
for i in genre:
    for j in range(len(i)):
        for k in py['genres']:
            if (i[j]==k['id']):
                i[j]=k['name']
for i in range(len(title)):
    print(title[i],'-', end = "")
    temp = len(genre[i])
    for j in range(temp):
        if (j!=temp-1):
            print("",genre[i][j],end=',')
        else:
            print("",genre[i][j], end="")
    print()
```

Fetch the trending TV Shows for the week from the TMDb API and print the taglines of the top 5 shows. If there is no tagline, print 'Empty' instead

```python
url = 'https://api.themoviedb.org/3/trending/tv/{time_window}'
url = url.format(time_window='week')

r = requests.get(url, headers=headers)
py = r.json()
```

```python
for i in py['results'][:5]:
    url_series = 'https://api.themoviedb.org/3/tv/{series_id}'
    url_series = url_series.format(series_id = i['id'])
    r = requests.get(url_series, headers=headers)
    py = r.json()
    if (py['tagline']==''):
        print('Empty')
    else:
        print(r.json()['tagline'])
```

Count the number of males and females in the cast of "Money Heist" using the TMDb API.

```python
r = requests.get('https://api.themoviedb.org/3/search/tv', headers=headers,
                params = {'query':'Money Heist'})
_id = r.json()['results'][0]['id']
```

```python
#male=2, female=1
url = 'https://api.themoviedb.org/3/tv/{series_id}/credits'
url = url.format(series_id=_id)

r = requests.get(url, headers=headers)
py = r.json()

male_count, female_count = 0, 0
for i in py['cast']:
    if (i['gender']==2):
        male_count+=1
    else:
        female_count+=1

print(male_count, female_count)
```
